Citric experiment A-F (2A-F) are from Januaury ICP-OES measurements, and G is from February measuremnt. 2F from February is not used, as there are significantly lower Al, Ca, Fe, Si, Mn, Sr, slightly lower S, Mg, and slightly higher K.

Citric experiment samples A-G (2A-G) are high in Na and over-saturated for OES to measure. Dilutions (1:10) were made and Na data are from the measurements of these dilutions multiplied by 10.

Citric-half control A (3A) was likely contaminated (4-127 times more Ca, Fe, Cu, Mn, Pb, and Zn than the other citric-matrix controls) and had been substituted with the data at B (the same control that was sampled 23 hours
later).

Citric-half experiment A (4A) was likely contaminated as well (high Zn, 150 times more Zn than citric experiment A, which had double the molarity of citric acid). However, it did not seem to be contaminated for other elements (with some element concentrations at A lower than those at B). Thus, only Zn data was made NaN for citric-half experiment at A.

On the other hand, HCl control C (11C) could be contaminated too, having 4-13 times more Zn, Ni, Mn, Cu, and Co compared to other HCl controls, and has been substituted with the average of these elements in HCl control B and D.

```python
import matplotlib.pyplot as plt
import numpy as np

# remove 'time' column
columns_to_drop = ['time']
citric_trele = citric_ms.drop(columns=columns_to_drop)
HCl_trele = HCl_ms.drop(columns=columns_to_drop)
oxalic_trele = oxalic_ms.drop(columns=columns_to_drop)
maleic_trele = maleic_ms.drop(columns=columns_to_drop)

# subsitute 3A for 3B
citric_trele.iloc[14] = citric_trele.iloc[15]

## for Zn, make 4A data NaN
citric_trele['Zn']['4A'] = np.nan

# Convert columns to numeric
HCl_trele = HCl_trele.apply(pd.to_numeric, errors='coerce')
# substitude 11C  Zn, Ni, Mn, Cu, Co for the average of concentrations at B and D
HCl_trele['Zn']['11C'] = np.mean([HCl_trele['Zn']['11B'],HCl_trele['Zn']['11D']])
HCl_trele['Ni']['11C'] = np.mean([HCl_trele['Zn']['11B'],HCl_trele['Zn']['11D']])
HCl_trele['Mn']['11C'] = np.mean([HCl_trele['Zn']['11B'],HCl_trele['Zn']['11D']])
HCl_trele['Cu']['11C'] = np.mean([HCl_trele['Zn']['11B'],HCl_trele['Zn']['11D']])
HCl_trele['Co']['11C'] = np.mean([HCl_trele['Zn']['11B'],HCl_trele['Zn']['11D']])

# calculate Delta Element (experiment minus control)
citric_ms_abs = np.array(citric_trele['2A':'2G'])-np.array(citric_trele['1A':'1G'])
citric_half_ms_abs = np.array(citric_trele['4A':'4G'])-np.array(citric_trele['3A':'3G'])

HCl_ms_abs = HCl_trele.iloc[7:14].values - HCl_trele.iloc[0:7].values

oxalic_ms_abs = np.array(oxalic_trele[7:14])-np.array(oxalic_trele[0:7])

maleic_ms_abs = np.array(maleic_trele[7:14])-np.array(maleic_trele[0:7])
```

## **converting for Sr and Mn**



```python
# to convert Sr from ppm to ppb
# data are all from ICP-OES

citric_Sr = citric_abs[:,10]*1000
citric_half_Sr = citric_half_abs[:,10]*1000
oxalic_Sr = oxalic_abs[:,8]*1000
maleic_Sr = maleic_abs[:,8]*1000
hcl_Sr = HCl_abs[:,5]*1000

# save Sr data into a dataframe
data_Sr = {
     "citric": citric_Sr,
     "citric_half": citric_half_Sr,
     "oxalic": oxalic_Sr,
     "maleic": maleic_Sr,
     "hcl": hcl_Sr
}

index = range(len(citric_Sr))
# Create the DataFrame with specified row names
Sr = pd.DataFrame(data_Sr, index=index)


# to convert Mn data in ppb
# using ICP-OES data for citric (convert this from ppm to ppb)
# using ICP-MS data for Maleic, Oxalic, HCl
citric_Mn = citric_abs[:,9]*1000
citric_half_Mn = citric_half_abs[:,9]*1000

data_Mn = {
     "citric": citric_Mn,
     "citric_half": citric_half_Mn
 }

index = range(len(citric_Sr))
# Create the DataFrame with specified row names
Mn = pd.DataFrame(data_Mn, index=index)
```
